What are the actions you can do when you deal with duplicates in the data frame

```python
# Create a copy of the data frame to use

dupi_df = df.copy()
```

Check if there are any duplicates in the dataframe

You are able to check if there are any duplicates in this dataframe by using the function .duplicated() and .sum()

```python
# Check how many duplicates there are in the dataframe
dupi_df.duplicated().sum()
```

Checking how many times a value shows up in the dataframe to find which are duplicated

You can count the amount of values in a column by using the function .value_counts()

```python
dupi_df['student_id'].value_counts()
```

Then we need to find where the duplicates are located in dateframe

We are able to find where the duplicates are gonna be located by using the .duplicated() function

```python
# Finding the exact location of where the duplicates are
dupi_df[dupi_df.duplicated()]
```

If we want to include the first of the duplicates as well we can use the keep argument to equal False

```python
# Finding the duplicates and the first occurance as well
dupi_df[dupi_df.duplicated(keep=False)]
```

Next we have to remove the duplicates

We can do this in multiple different ways, how it is done depends on what you need done

1 We can just all of the exact rows that are duplicates that are identical

Dropping all of the exact rows that are duplicates you need to use the function .drop_duplcates()

```python
# Use the .drop_duplicates to remove all of the duplicates
dupi_df.drop_duplicates()
```

Dropping by a singular coplumn or mulitple columns

Dropping by specific columns is done by using the drop_duplicates function but you pass the subset of either a column name or a list of column names

```python
dupi_df.drop_duplicates(subset = "student_id")
```

Keep the last duplicate instead of the first

This is done by passing in another argument in the function of .drop_duplicates(), the argument is keep where it will equal "last"

```python
# Dropping the duplicates in the student_id column
# but keeping the last insetead of the first
dupi_df.drop_duplicates(subset = "student_id", keep = False)
```
