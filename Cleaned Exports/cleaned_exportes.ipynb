```python
# Imports that are used
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    "employee_id": [101, 102, 103, 104, 105],
    "employee_name": ["John Smith", "Maria Lopez", "Kevin Brown", "Aisha Khan", "Daniel Lee"],
    "department": ["Sales", "Marketing", "IT", "Human Resources", "Finance"],
    "salary": [55000, 62000, 58000, 72000, 69000],
    "hire_date": pd.to_datetime(["2022-01-15", "2021-07-10", "2023-03-22", "2020-11-05", "2019-09-18"]),
    "full_time": [True, False, True, True, True]
})

df
```

Checking befoe exporting

Need to check before exporting the dataframe to ensure that all of the data is in order. This is done by either .info() or .head()

```python
# Create a copy of the data frame
export_df = df.copy()
```

```python
# Use the .info()
export_df.info()
```

```python
# Use the .head()
export_df.head()
```

Exporting the cleaned Data Frames

1 Export to CSV

```python
# Export the data frame to csv
export_df.to_csv("cleaned_employee_data.csv", index=False)
```

2 Export to Excel

```python
# Export the dataframe to Excel
export_df.to_excel("cleaned_employee_data.xlsx", index=False)
```

3 Export to JSON 

```python
# Export the dataframe to JSON
export_df.to_json("cleaned_employee_data.json", orient="records", indent=4)
```

4 Export to Parquet

```python
# Export the data into Parquet
export_df.to_parquet("cleaned_employee_data.parquet", index=False)
```

5 Export selected columns only

```python
# Create a list of selected columns_df
selected_cols = export_df[["employee_id", "employee_name", "department", "salary"]]
# Export the selected colums df into csv
selected_cols.to_csv("employee_summary.csv", index = False) 
```

6 Export filtered data only

```python
# Use boolean masking to filter the dataframe
full_time_df = export_df[export_df["full_time"] == True]
# Export the saved dataframe into the CSV
full_time_df.to_csv("full_time_employees.csv", index=False)
```

7 Export Sorted Data

```python
# Sort the dataframe by the sales column
sorted_df = export_df.sort_values(by="salary", ascending=False)
# Export the sorted dataframe into csv
sorted_df.to_csv("employees_sorted_by_salary.csv", index=False)
```

8 Export with cleaned data formatting

```python
# Create a formatted date colunmn
export_df["hire_date_clean"] = export_df["hire_date"].dt.strftime("%Y-%m-%d")
# Saved the formated the dataframe to csv
export_df.to_csv("cleaned_employee_data_dates.csv", index=False)
```
