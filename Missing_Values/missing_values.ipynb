```python
# Import the neccessary libraries
import pandas as pd
import numpy as np
```

```python
# Creating the the messy synthetic data
import pandas as pd
import numpy as np

data = {
    "Student": ["Alex", "Jordan", "Taylor", "Morgan", "Casey", "Riley"],
    "Math_Score": [85, np.nan, 78, 92, np.nan, 88],
    "English_Score": [90, 76, np.nan, 85, 80, np.nan],
    "Attendance_Rate": [0.95, 0.88, np.nan, 0.97, 0.91, np.nan]
}

df = pd.DataFrame(data)

df
```

1 Dropping the NA

The first way to handle the missing data is just to drop it using the .dropNA() function

```python
# See before using the .dropNA()
drop_df = df.copy()
drop_df.head()
```

```python
# Using the dropNA() to drop the NA 
# from the dataframe

drop_df = drop_df.dropna()
drop_df
```

2 Fill the Numeric Values with a Median

The second way to handle the missing data is to fill in the numeric missing data with the median using the .fillna() function


```python
# Create the fill_df as a copy of the df
# To show before
fill_df = df.copy()
fill_df.head()
```

```python
# Use the .fillna function
fill_df['Math_Score'] = fill_df['Math_Score'].fillna(fill_df['Math_Score'].median())
fill_df['English_Score'] = fill_df['English_Score'].fillna(fill_df['English_Score'].median())
fill_df['Attendance_Rate'] = fill_df['Attendance_Rate'].fillna(fill_df['Attendance_Rate'].median())
```

```python
# The result after using the second method
fill_df
```

3 Fill the string NAs with labels

The third way to get rid of the NAs is to fill them with labels. This is done also by using the .fillna() function.

```python
# Create a new df that has strings for values
df2 = pd.DataFrame({
    "employee_id": [101, 102, 103, 104, 105, 106, 107, 108],
    "department": [
        "Sales",
        np.nan,
        "Finance",
        "HR",
        np.nan,
        "IT",
        "Marketing",
        np.nan
    ],
    "job_title": [
        "Analyst",
        "Manager",
        np.nan,
        "Coordinator",
        "Analyst",
        np.nan,
        "Director",
        "Manager"
    ],
    "employment_type": [
        "Full-Time",
        "Part-Time",
        np.nan,
        "Contract",
        "Full-Time",
        np.nan,
        "Intern",
        "Full-Time"
    ]
})

df2
```

```python
string_df = df2.copy()
string_df.head()
```

```python
string_df['department']
```

```python
# Using the final method the replace the NA
string_df['department'] = string_df['department'].fillna("Not known")
string_df['job_title'] = string_df['job_title'].fillna("Not known")
string_df['employment_type'] = string_df['employment_type'].fillna("Not known")
```

```python
# The result of the third method
string_df
```
