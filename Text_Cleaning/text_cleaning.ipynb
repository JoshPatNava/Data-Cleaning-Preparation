Cleaning Texts

There are many things that can be done to clean texts.

```python
# Create a copy of the data frame
text_df = df.copy()
text_df.head()
```

1 Remove the Spaces at the beginning and the end

Removing the spaces at the beginning and the end is possible by using the function of .str.strip() on a column

```python
# Remove the end and beginning spaces for the columns name, email, and city
text_df['name'] = text_df['name'].str.strip()
text_df['email'] = text_df['email'].str.strip()
text_df['city'] = text_df['city'].str.strip()

# Check the new data frame
text_df.head()
```

2 Change text to lowercase

Chaning the text values to lowercase is done by using fucntion str.lower() on that column

```python
# Change both the name and the city to lower case

text_df['name'] = text_df['name'].str.lower()
text_df['email'] = text_df['email'].str.lower()

text_df.head()
```

3 Change text to title case

Changing the text to the title case is done by the function of .str.title()

```python
# Change the names and the city columns into title case
text_df['name'] = text_df['name'].str.title()
text_df['city'] = text_df['city'].str.title()
# Check the dataframe
text_df.head()
```

4 Raplace characters

In order to replace characters in a column that have text in it you need to use the function of .str.replace()

```python
# Replace both the - and the _ in the names columns
text_df['name'] = (
    text_df['name']
    .str.replace("_"," ", regex=False)
    .str.replace("-", " ", regex=False)
)

# Check the data frame
text_df.head()
```

5 Remove extra spaces inside text

Removing the extra spaces that are inside the text uses the same function but you pass the argument of r"\s+" for the first one 

```python
# Remove the extra spacing in both the columns 
# of the name and the city
text_df['name'] = text_df['name'].str.replace(r"\s+", " ", regex = False)
text_df['city'] = text_df['city'].str.replace(r"\s+", " ", regex = False)
text_df.head()
```

6 Remove punctuation or symbols

Removing the punctuation or the symbols within a column with texts you need to use the .str.replace() and pass the argument r"[^a-zA-Z\s]", " ", regex = True

```python
# Remove the punctuations or the symbols in the city column
text_df["city"] = text_df["city"].str.replace(r"[^a-zA-Z\s]", "", regex=True)
# Check the data frame
text_df.head()
```

7 Replace placeholder values with missing values

Replace the placeholder values such as N/A, NA, na, etc, with a an actual np.nan. Using the .str.replace()

```python
# Replace all of ht placeholder values with actual np.nan
text_df = text_df.replace(["N/A", "n/a", "NA", "na", "unknown", "Unknown"], np.nan)
# Check the data frame
text_df.head()
```

8 Clean phone numbers

Cleaning the phone numbers utilizes the same function of .str.replace(), except you are passing in the argument of r"\D". Which basically removes everything and just leaves the numbers

```python
# Keep only the numbers within the column for numbers
text_df['phone'] = text_df['phone'].str.replace(r"\D", "", regex = True)
# Check the dataframe
text_df.head()
```

9 Format phone number after cleaning 

For further cleaning of the the phone numbers you can also add some formatting to the number, by using the same function of .str.replace()

```python
# Refromat the phone number to match a common phone number
text_df["phone"] = text_df["phone"].str[-10:]

text_df["phone_formatted"] = (
    "(" + text_df["phone"].str[0:3] + ") " +
    text_df["phone"].str[3:6] + "-" +
    text_df["phone"].str[6:10]
)

# Check the dataframe
text_df.head()
```

10 Check whether text contains something

To be able to check whether a text contains something you can use the function of .str.contains()

```python
# Check to see if there 
# are any gmails within the email column
text_df['is_gmail'] = text_df['email'].str.contains("gmail", na=False)
# Check the dataframe
text_df.head()
```

11 Check if emails look valid

Checking if the emails are valid in the email column you need to be able to use the .str.contains() function and pass the regex argument of r"^[\w\.-]+@[\w\.-]+\.\w+$"

```python
# Check if the emails in the emails column 
# would be considerd as valid
text_df['valid_email'] = text_df['email'].str.contains(
    r"^[\w\.-]+@[\w\.-]+\.\w+$",
    regex = True,
    na = False
)

# Checking the dataframe
text_df.head()
```

12 Split text into new columns

To split the text that are in columns you need to use the function of .str.split()

```python
# In the name column split the text 
# into the first name and last name
text_df[['first_name', 'last_name']] = text_df['name'].str.split(
    " ",
    n=1,
    expand = True
)

# Check data frame
text_df.head()
```

13 Extract text using regex

Extracting text uses the .str.extract() function and to be able to use it you need to add the regex argument as well

```python
# Extract the email domain within the email column
text_df["email_domain"] = text_df["email"].str.extract(r"@([\w\.-]+)")
# Check the data frame
text_df.head()
```

14 Count text length

Counting the text length within a text column uses the function of .str.len()

```python
# Find the text length of the emails
text_df['email'].str.len()
```

15 Chain text cleaning steps together

Finally you are able to chain all of these function and methods together to clean a whole column in a single go.

```python
# Recopy the dataframe
new_df = df.copy()
```

```python
# Completely clean the name column.
new_df["name_clean"] = (
    new_df["name"]
    .str.strip()
    .str.replace("_", " ", regex=False)
    .str.replace("-", " ", regex=False)
    .str.replace(r"\s+", " ", regex=True)
    .str.title()
)

new_df[['name_clean', 'name']]
```
