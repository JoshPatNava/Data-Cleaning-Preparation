```python
# Imports that we are gonna be using
import pandas as pd
import numpy as np
```

```python
# Creating the fake data 
import pandas as pd

data = pd.DataFrame({
    "student": ["Alex", "Bella", "Chris", "Diana", "Ethan", "Fiona", "George", "Hannah", "Ivan", "Jasmine"],
    "major": ["CS", "Business", "CS", "Nursing", "Business", "CS", "Nursing", "Business", "CS", "Nursing"],
    "grade_level": ["Freshman", "Sophomore", "Junior", "Senior", "Freshman", "Senior", "Junior", "Sophomore", "Freshman", "Senior"],
    "exam_score": [88, 72, 95, 64, 81, 90, 58, 76, 69, 84],
    "attendance": [92, 85, 98, 70, 88, 95, 60, 80, 75, 91],
    "hours_studied": [6, 3, 9, 2, 5, 8, 1, 4, 3, 7],
    "passed": [True, True, True, False, True, True, False, True, True, True]
})

data
```

Filtering Methods

```python
# Copying the data into a seperate data frame
filter_df = data.copy()
# Check the data
filter_df.head()
```

Filtering using 1 Condition

Using only 1 condition to filter out from a column

```python
# Going to filter 
# on students that have an exam 
# score of more than 75

more_than_75 = filter_df[filter_df['exam_score'] > 75]
more_than_75
```

Filtering mutiple conditions using &

Using multple condition on the same column or different you need to use the ampersand to link the filtering conditions togther.

```python
# We are gonna filter by students that have an exam 
# score less than or equal to 60 
# and their attendance is below 55

less_60_below_55 = filter_df[(filter_df['exam_score'] <= 75) & (filter_df['attendance'] < 80)]
less_60_below_55
```

Filter using the | for the OR

Filter using the |(OR) allows the condition to be either one or the other. 

```python
# We want to filerwhether a student 
# has an example score of above 85 
# or they studied for 5 hours

or_method = filter_df[(filter_df['exam_score'] > 85) | (filter_df['hours_studied'] > 5)]
or_method
```

Filter using the !=

Using the != to filter out a specific value in the dataframe

```python
# Filtering for the student who did not pass

not_method = filter_df[filter_df['passed'] != True]
not_method
```

Filter using the function of .isin()

Using the .isin() functions takes a list and checks whether values in the list are in the column that you specified.

```python
# Filtering to see whether students are in Nursing

isin_method = filter_df[filter_df['major'].isin(["Nursing"])]
isin_method
```

Filter using the function of .between()

Using the .between() function takes two values and filters all of the rows between those values, will most likely be a numerical value.

```python
# Filtering to find the students exam score
# that are between the values of 65 and 85

between_method = filter_df[filter_df['exam_score'].between(65, 85)]
between_method
```

Filter using the string methods, using str.startswith()

Filtering string columns using the .str.startswith() allows to check columns that have a string value and find whether they start with something

```python
# Finding Students names that start with the letter B

start_method = filter_df[filter_df['student'].str.startswith("B")]
start_method
```

Filter using the .query() function

Another way to query, and filter the dataframe, it is preference but may sometimes look cleaner

```python
# Filter the students that have the exam score of 
# more than 80 and attendance more than 90

query_method = filter_df.query("exam_score >= 80 and attendance >= 90")
query_method
```

Sorting Methods

```python
# Create a copy of the dataframe for the sorting methods

sort_df = data.copy()
```

Sorting by one column

Using only a singular column to sort the whole data frame, we are able to do this by using the function .sort_values()

```python
# Sort the dataframe by the students exam score in ascending order

lowest_score =  sort_df.sort_values(by = ['exam_score'])
lowest_score
```

Sorting the columns by decending

Using the same function of .sort_values(), if we add the parameter of the acending=False, then it will order the data frame by descedning order based on the column that we are focusing on.

```python
# Sort the dataframe by the the exam scores but in the descending order.

high_score = sort_df.sort_values(by = ['exam_score'],ascending=False)
high_score
```

Sorting by the multiple columns

Sorting using multiple columns you have to use the & symbol in order to connect the conditions, by adding more columns into to the list.

```python
# Sort the dataframe by the columns exam score and the attendance.

exam_attendance = sort_df.sort_values(by = ['exam_score', 'attendance'])
exam_attendance
```

Sort multiple columns with different directions

In addition to sorting by multiple columns you are able to sort them with the condition that they will be going into different directions. By passing in the a list of true and falses in the ascending variable

```python
# Sort the dataframe by attendance and hours studied, first in descending order and then the second in ascending order
attendance_hours = sort_df.sort_values(by = ['attendance', 'hours_studied'], ascending = [False, True])
attendance_hours
```

Sort by index

Sorting by the index you are able to do this by using the function sort_index()

```python
# Looking at the one of the previous sort dataframes that we saved
attendance_hours.head()
```

```python
# Sort by the index 
attendance_hours.sort_index()
```
